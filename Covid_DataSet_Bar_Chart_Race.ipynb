```python
import pandas as pd
```

```python
df=pd.read_csv("corona_dat.csv")
```

```python
df.head()
```

```python
df2=df[["date","China","Italy","Brazil","Spain","US","United Kingdom","France","Turkey"]]
```

```python
df.info()
```

```python
df2.set_index("date",inplace=True)
```

```python
df2
```

```python
df_total=df2.cumsum(axis=0)
```

```python
df_total.head()
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(df_total,filename="covid19.mp4",figsize=(8,5),title="İbrahim - Covid19 Verileri")
```
