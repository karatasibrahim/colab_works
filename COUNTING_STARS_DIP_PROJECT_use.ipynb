Importing the image  and convering it into grayscale image 

```python
Starimage=cv2.imread('/content/drive/MyDrive/counting star111.png')

cv2_imshow(Starimage)

grayscaled_star_image =cv2.cvtColor(Starimage, cv2.COLOR_RGB2GRAY)
x1,y1=(np.shape(grayscaled_star_image)) 
print(x1,y1)
```

```python
k=0
H=np.zeros((256), dtype=int)
#print(H)
while k<256:
  H[k]=np.count_nonzero(grayscaled_star_image==k)
  k+=1
Intensity=np.arange(0,256,1)
#print(Intensity)
#print(H)
plt.bar(Intensity,H, color='red', width=1.0)
#plt.ylim(0,2000)
plt.xlabel("intensity")
plt.ylabel("frequency")
plt.title("histogram")

plt.show
```

```python
'''Thresholdin the image that is selecting intensity values for a specific range '''

x,y = np.shape(grayscaled_star_image)
binary_image= np.zeros((x,y))
for p in range(x):             
  for q in range(y):
    if (grayscaled_star_image [p][q]<120):  
      binary_image[p][q]=0
    else:
      binary_image[p][q]=255


cv2_imshow(binary_image)             #printing the binary image


s=binary_image
xsize ,ysize= np.shape(s)

count=0
for p in range(xsize):
    for q in range(ysize):
      if s[p][q]==255:
        count = count+1
        s[p][q]=0
        if p==x1 or q==y1:
          s[p][q]=0
        elif p==x1-1 or q==y1-1 :
          s[p][q]=0
        elif p==x1-2 or q==y1-2: 
          s[p][q]=0
        elif p==0 or q==0: 
          s[p][q]=0
        elif p==1 or q==1: 
          s[p][q]=0

    
        else:
          for i in range(-2,3):
            for j in range(-2,3):
              a=s[p+i][q+j] = 0            
              

'''
          s[p+1][q] = 0 
          s[p-1][q] = 0 
          s[p][q-1] = 0
          s[p][q+1] = 0
          s[p+1][q-1] = 0
          s[p-1][q-1] = 0
          s[p+1][q+1] = 0
          s[p-1][q+1] = 0
          
          s[p+2][q] = 0
          s[p-2][q]=0
          s[p][q-2] = 0
          s[p][q+2] = 0
          s[p-1][q-2] = 0
          s[p-2][q-2] = 0
          s[p-2][q-1]=0
          s[p+1][q-2]=0
          s[p+2][q-2]=0
          s[p+2][q-1]=0
          s[p-2][q+1] = 0
          s[p-2][q+2]=0
          s[p-1][q+2]=0
          s[p+1][q+2]=0
          s[p+2][q+1]=0
          s[p+2][q+2]=0
'''     
    

print('The total star count =', count)
#cv2_imshow(s)
```
